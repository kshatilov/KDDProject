```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.cluster as cluster
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.decomposition import FastICA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import time
%matplotlib inline

sns.set_context('poster')
sns.set_color_codes()
plot_kwds = {'alpha' : 0.25, 's' : 80, 'linewidths':0}
```

## 1. PCA

```python
data_wine = datasets.load_wine()

X = data_wine.data
y = data_wine.target
target_names = data_wine.target_names

pca = PCA(n_components=5)
X_pca = pca.fit(X).transform(X)
print (pca.explained_variance_ratio_)
print ("sum : %s" % str(sum(pca.explained_variance_ratio_))  )
```

```python
pca = PCA(n_components=1)
X_pca = pca.fit(X).transform(X)
print (pca.explained_variance_ratio_)
print ("sum : %s" % str(sum(pca.explained_variance_ratio_))  )
```

## 2. FastICA

```python

num_classes = len(target_names)
ica = FastICA(n_components=num_classes)

X_ica = ica.fit(X).transform(X)
```

```python
print (X_ica)
```

## 3. LDA

```python
lda = LinearDiscriminantAnalysis(n_components=num_classes)
X_lda = lda.fit(X, y).transform(X)
```
